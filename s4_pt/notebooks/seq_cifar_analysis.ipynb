```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open("../artifacts/seq_cifar/loss.json") as f:
    loss = np.array(json.load(f))
```

```python
window = 50
smooth_loss = np.convolve(loss, np.ones(window)/window, mode="valid")

plt.figure(figsize=(6,4))
plt.plot(smooth_loss)
plt.axhline(np.log(10), color="r", linestyle="--", label="Random baseline")
plt.xlabel("Training step")
plt.ylabel("Cross-entropy loss")
plt.title("Sequential CIFAR-10 â€“ S4D Training Loss")
plt.legend()
plt.grid(True)
plt.show()
```

```python
print("Min loss:", smooth_loss.min())
print("Final loss:", smooth_loss[-1])
```

```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open("../artifacts/speech/loss.json") as f:
    loss = np.array(json.load(f))
```

```python
plt.figure(figsize=(6,4))
plt.plot(loss)
plt.xlabel("Training step")
plt.ylabel("Cross-entropy loss")
plt.title("Speech Commands â€“ S4D Training Loss")
plt.grid(True)
plt.show()
```

```python
print("Initial loss:", loss[:10].mean())
print("Final loss:", loss[-10:].mean())
```
